```python
import torch
import torch.nn as nn
import torch.nn.functional as F

def relu(x):
    return max(0, x)

class DeepModel(nn.Module):
    def __init__(self, n_features, n_hidden, n_hiddentwo, n_outputs):
        super().__init__()
        self.fc1 = nn.Linear(n_features, n_hidden)
        self.fc2 = nn.Linear(n_hidden, n_hiddentwo)
        self.fc3 = nn.Linear(n_hiddentwo, n_outputs)

    def forward(self, x):
        z = F.sigmoid(self.fc1(x))
        ztwo = F.sigmoid(self.fc2(z))
        output = F.sigmoid(self.fc3(ztwo))
        # z = F.relu(self.fc1(x))
        # ztwo = F.relu(self.fc2(z))
        # output = F.relu(self.fc3(ztwo))
        return output

# Create the model with the correct number of input layer neurons, a hidden layer, and the correct number of output neurons
model = DeepModel(n_features = 12, n_hidden = 256, n_hiddentwo = 100, n_outputs=3)

# Example training data (random)
X_train_tensor = torch.rand(10, 12)
# Get the model outputs
outputs = model(X_train_tensor)

# Find the mean values for each output neuron
mean_outputs = torch.mean(outputs, dim=0)

# Find the output neuron with the largest value
max_value, max_index = torch.max(mean_outputs, dim=0)

print("Outputs:\n", mean_outputs)
print("Neuron with largest value is at index:\n", max_index)
```
