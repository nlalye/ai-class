```python
m=4
features = np.array([1.0,2.0,3.0,4.0])
actual = np.array([55, 60, 66, 71])
thetaZero = 50
thetaOne = 5

predicted = thetaZero + thetaOne * features

print("Y_hat:", predicted)
```

```python
errors = predicted - actual

J = np.dot(errors, errors) / (2 * m)
print("Initial Cost:", J)
```

```python
learning_rate = 0.01

thetaZero -= learning_rate * (1/m) * np.sum(errors)
thetaOne -= learning_rate * (1/m) * np.sum(errors * features)

print("Theta0:", thetaZero)
print("Theta1:", thetaOne)

predicted = thetaZero + thetaOne * features
errors_new = predicted - actual
J_new = np.dot(errors_new, errors_new) / (2 * m)
print("New Cost:", J_new)

    
```

```python
import matplotlib.pyplot as plt

plt.scatter(features, predicted, color='purple', marker = 'o', label='Data points')
plt.plot(features, predicted)

plt.xlabel('Study Hours')
plt.ylabel('Grades')
plt.title('Scatter Plot with Line of Best Fit')
plt.legend()
plt.grid(True)
plt.show()
```
